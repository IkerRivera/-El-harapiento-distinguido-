# Las variables
1. CT =  Costo del traje
2. DE =  Descuento que se obtendrá
3. PF = Precio final del traje

# formulas
* DE = CT * 0.15
* DE = CT * 0.08
* PF = CT – DE

# diagrama de flujo lucidchart
<img src="https://raw.githubusercontent.com/IkerRivera/-El-harapiento-distinguido-/refs/heads/main/DF%20Traje%204.png">

```python
# Entrada de datos
# # para cuando es verdadero (si)
# calculando el cambio usando if else
# Datos de entrada DE , PF
# mostar descuento mas de 2500
CT = float(input("Ingrese el costo del traje: "))

if CT > 2500:
    DE = CT * 0.15
else:
    DE = CT * 0.08

PF = CT - DE

print("El descuento es: ", DE)
print("El precio final a pagar es: ", PF)

print("Fin del programa")
print("Programa realizado por Iker Montoya NL = 35")
```
